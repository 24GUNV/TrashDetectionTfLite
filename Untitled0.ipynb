Splitting the data into train / validate / test sets




```python
!git clone https://github.com/24GUNV/aibuilders.git
!rm -r aibuilders/ai
!mkdir aibuilders/ai
!mkdir aibuilders/ai/valid
!mkdir aibuilders/ai/test
!mkdir aibuilders/ai/train

import glob
import shutil
import os

names = {
    0 : "cardboard",
    1 : "glass",
    2 : "metal",
    3 : "paper",
    4 : "plastic",
    5 : "trash"
}


valid_folders = glob.glob('aibuilders/images/*')
valid_fnames = []
for folder in valid_folders:
  valid_fnames.append(glob.glob(f'{folder}/*'))

for i in range(6):
  os.mkdir(f'aibuilders/ai/valid/{names[i]}')
  os.mkdir(f'aibuilders/ai/test/{names[i]}')
  os.mkdir(f'aibuilders/ai/train/{names[i]}')

for folder in valid_fnames:
  for i, fname in enumerate(folder):
    if i < int(len(folder)*0.25): # Takes 25% and puts in validation set
      shutil.copyfile(fname, f'aibuilders/ai/valid/{fname.split("/")[-2]}/{fname.split("/")[-1]}')
    elif i < int(len(folder)*0.9): # Takes 10% and puts it in testing set
      shutil.copyfile(fname, f'aibuilders/ai/train/{fname.split("/")[-2]}/{fname.split("/")[-1]}')
    else: # Last 10% is put into the testing set
      shutil.copyfile(fname, f'aibuilders/ai/test/{fname.split("/")[-2]}/{fname.split("/")[-1]}')
```

Loading images into separate batches of 15


```python
from fastbook import *

# Idk what going on here just copied and pasted from lesson
dblock = DataBlock(
    blocks=(ImageBlock, CategoryBlock), #x - image; y - single class
    get_items=get_image_files, #get image
    splitter=GrandparentSplitter(valid_name='valid_mini'), #use parent folder as train-valid split
    get_y=parent_label, #use parent folder as label
    batch_tfms=aug_transforms(size=224)
    )
dls = dblock.dataloaders('aibuilders/ai/', bs=15)
```

Testing to see if the batch laoded in correctly


```python
dls.train.show_batch(max_n=9,nrows=3)
```

AI actually learning

```python
learn = cnn_learner(dls, resnet34, metrics=accuracy)
learn.fine_tune(epochs=0, freeze_epochs=1, base_lr=2e-3)
```

```python
learn.show_results() #true label - บน; prediction - ล่าง
```
